```python
#Dataframe
#las claves son las columnas
data={
    "Nombre": ["Ana", "Juan", "Pedro", "Maria","Luis"],
    "Edad": [20,23,28,50,40],
    "Ciudad": ["Madrid", "Barcelona", "Valencia", "Sevilla","Malaga"]
}
df=pd.DataFrame(data)
df, type(df)
```

```python
#data frame a partir de series
df2=pd.DataFrame(data=serie)
df2
```

```python
#exportar datos
df.to_csv("datos.csv") #nombre con el que se va a guardar el archivo
```

```python
#manipulacion de datos
#Seleccionar una columna
nombre=df["Nombre"]
nombre
```

```python
#seleccionar varias columnas
df[["Nombre","Edad"]]
```

```python
#Filtrar por indixes
fila = df.loc[0]
fila
```

```python
#filtrar por condiciones
df[df["Edad"]>25]
```

```python
filtro= (df["Edad"]>23) & (df["Nombre"].str.startswith("P"))
df[filtro]
```

```python
#filtrar por query
df.query("Edad>23")
```

```python
df[df["Nombre"].isin(["Ana","Carlos","Luis"])]
```

```python
#funciones
def longitud(nombre):
  return len(nombre)==5

df[df["Nombre"].apply(longitud)]
```

```python
#filtrar por edades(inclusivo)
df[df["Edad"].between(28,40)]
```
